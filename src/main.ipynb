```python
respuestas = []

for solicitud in solicitudes:
    # Llamar al modelo
    respuestas.append(json.loads(agente.onetime_query_model(solicitud)))
```

```python
'''
Imprimir como tabla en pantalla las respuestas del modelo
'''
from IPython.display import display, Markdown

# Encabezado de la tabla
md_table = "| Opener | Confianza | Chain of Thought |\n"
md_table += "|--------|-----------|----------------|\n"

for respuesta in respuestas:
    if 'Error' in respuesta.keys():
        # En caso de error, rellenar y skip
        opener = 'Error'
        confianza = 0 
        chain = 'Error' 
        md_table += f"| {opener} | {confianza} | {chain} |\n"
        continue

    # Escapar saltos de línea y pipes dentro del contenido
    opener = respuesta['opener'].replace("\n", " ").replace("|", "\\|")
    confianza = respuesta['confianza']
    chain = " <br><br> ".join(respuesta['chain_of_thought']).replace("|", "\\|")

    md_table += f"| {opener} | {confianza} | {chain} |\n"

# Mostrar en Jupyter como Markdown
display(Markdown(md_table))
```
